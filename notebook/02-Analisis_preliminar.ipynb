### Columnas del dataset

● ``order_id``: Identificador único del pedido realizado.

● ``order_date``: Fecha en la que se realizó el pedido.

● ``ship_date``: Fecha en la que el pedido fue enviado al cliente.

● ``ship_mode``: Tipo de envío utilizado (por ejemplo: entrega estándar, rápida o el mismo día).

● ``customer_name``: Nombre del cliente que realizó la compra.

● ``segment``: Segmento de mercado al que pertenece el cliente (por ejemplo: consumidor, empresa, hogar).

● ``state``: Estado o provincia donde se realizó el pedido.

● ``country``: País donde se encuentra el cliente o se registró la venta.

● ``market``: Región comercial global (por ejemplo: Asia Pacífico, Europa, América Latina, etc.).

● ``region``: Subdivisión geográfica dentro del mercado (por ejemplo: norte, sur, centro).

● ``product_id``: Identificador único del producto vendido.

● ``category``: Categoría general del producto (por ejemplo: Tecnología, Muebles, Suministros de oficina).

● ``sub_category``: Subcategoría del producto (por ejemplo: Sillas, Teléfonos, Papel).

● ``product_name``: Nombre o descripción del producto.

● ``sales``: Valor total de las ventas de ese pedido (importe monetario).

● ``quantity``: Cantidad de unidades del producto vendidas.

● ``discount``: Descuento aplicado al producto en esa venta (en proporción, entre 0 y 1).

● ``profit``: Beneficio neto obtenido por la venta del producto (ventas menos costes).

● ``shipping_cost``: Coste del envío asociado al pedido.

● ``order_priority``: Nivel de prioridad asignado al pedido (por ejemplo: alta, media, baja).

● ``year``: Año en que se realizó el pedido.

● ``admin_name``: Nombre administrativo de la división territorial (estado o provincia).

● ``latitud``: Coordenada geográfica de latitud correspondiente al lugar del pedido.

● ``longitud``: Coordenada geográfica de longitud correspondiente al lugar del pedido.

● ``poblacion``: Población estimada de la ciudad o región donde se realizó el pedido.

● ``latitud_pais``: Coordenada geográfica de latitud correspondiente del país desde donde se hace el pedido.

● ``longitud_pais``: Coordenada geográfica de longitud correspondiente del país desde donde se hace el pedido.

● ``poblacion_pais``: Población estimada del país desde donde se hace el pedido.

### Observaciones rápidas

- ``quantity``: El mínimo es 1 y el máximo 14, lo cual es perfectamente razonable para pedidos minoristas. La media es 3.48, lo que indica que la mayoría de los pedidos contienen pocas unidades. La desviación estándar (2.28) muestra poca dispersión, es decir, las cantidades pedidas son consistentes.

- ``discount``: Rango entre 0 y 0.85, es decir, hay descuentos de hasta el 85%, algo alto pero posible en promociones. Media 0.14 (14%) y mediana 0, lo que sugiere que la mayoría de los pedidos no tienen descuento, pero algunos tienen descuentos muy grandes. Distribución muy sesgada a la derecha (muchos ceros, pocos descuentos extremos).

- ``profit``: El valor mínimo es -6599.98, indicando pedidos con grandes pérdidas. El máximo es 8399.98, con media 28.64 y mediana 9.24, lo que muestra gran dispersión. La desviación estándar es altísima (174.42), confirmando que hay outliers fuertes (beneficios o pérdidas excepcionales).

- ``shipping_cost``: Va de 0 a 933.57, con una media de 26.38 y mediana 7.79. Esto indica que la mayoría de los envíos son baratos, pero hay algunos envíos caros que elevan la media. Distribución también sesgada a la derecha.

- ``year``: Oscila entre 2011 y 2014, lo que confirma que el dataset cubre 4 años de operaciones. La media es 2012.78, lo que indica que la mayoría de los registros pertenecen a los años 2012–2013.

- ``latitud``: Rango: -46.23 a 62.04, lo cual es coherente con ubicaciones en el planeta Tierra (valores dentro del rango geográfico real). Media 24.22, lo que sugiere una concentración de pedidos en el hemisferio norte.

- ``longitud``: Rango: -122.54 a 178.02, también correcto para coordenadas reales. Media -4.45, indicando que los pedidos se ubican alrededor del meridiano de Greenwich, aunque con presencia global (occidente y oriente).

- ``poblacion``: Mínimo 0 y máximo 1,360,846,000, es decir, hay registros asociados tanto a ciudades pequeñas como a países muy poblados (por ejemplo, India o China). Media 22,747,452, lo cual tiene sentido: el promedio incluye muchas ciudades grandes y capitales. Desviación estándar muy alta (36,323,856), lo que confirma que hay una mezcla entre ciudades pequeñas y megaciudades.

- ``latitud_pais``: Rango: -40.63 a 62.04, lo cual es coherente con ubicaciones en el planeta Tierra (valores dentro del rango geográfico real). Media 24.13, lo que sugiere una concentración de pedidos en el hemisferio norte.

- ``longitud_pais``: Rango: -99.33 a 167.74, también correcto para coordenadas reales. Media -3.64, indicando que los pedidos se ubican alrededor del meridiano de Greenwich, aunque con presencia global (occidente y oriente).

- ``poblacion_pais``: Mínimo 110,000 y máximo 1,360,000,000, es decir, hay registros asociados tanto a ciudades pequeñas como a países muy poblados (por ejemplo, India o China). Media 181,846,600, lo cual tiene sentido: el promedio incluye muchas ciudades grandes y capitales.
```python
for col in columnas_num:
    plt.figure()
    sns.histplot(df[col])
    plt.title(f"Histograma de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.show()
```

### Countplot

```python
columnas_cat = df.select_dtypes(include=['category', 'object']).columns
```

```python
for col in columnas_cat:

    if df[col].nunique() > 200:
        print(f"Columna {col} tiene demasiadas categorías ({df[col].nunique()}\n\n")
        continue

    print(f"Distribución de la columna {col}")
    plt.figure(figsize=(df[col].nunique() * 2, 3))
    sns.countplot(x=df[col], order=df[col].value_counts().index)
    plt.show()
```

En los countplots podemos ver las distribuciones por columnas de las variables categóricas con sus diferentes valores.
